```python
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,20,30,40,50]
plt.plot(x,y, color = "yellow")
plt.title("simple line wala graph")
plt.xlabel("x-axis")
plt.ylabel("y-axis")

plt.show()
```

```python
import matplotlib.pyplot as plt
name = ["Ali","hamza", "Ahmed"]
marks = [40,80,75]

plt.bar(name, marks)
plt.title("Bar chart")
plt.show()
```

```python
#scatter ploter
import matplotlib.pyplot as plt
x = [1,2,3,4,5]
y = [10,20,30,40,50]

plt.scatter(x,y, color = "purple")
plt.title("Scatter plot")
plt.show()
```

```python
import matplotlib.pyplot as plt
marks = [10,40,20,60,80,75,99,25]

plt.hist(marks, color = "orange")
plt.title("Histogram")
plt.show()
```

```python
import matplotlib.pyplot as plt
subjects = ["physics", "chemistry", "computer"]
marks = [45,60,55]

plt.pie(marks, labels = subjects , autopct = "%1.1f%%")
plt.title("pie chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]
sales_2023 = [10,20,30,70]
sales_2024 = [40,10,50,60]

plt.plot(x, sales_2023, label = "2023 ki sales")
plt.plot(x, sales_2024, label = "2024 ki sales")
plt.title("multiline graph")
plt.xlabel("months")
plt.ylabel("sales")
plt.legend()
plt.show()
```
